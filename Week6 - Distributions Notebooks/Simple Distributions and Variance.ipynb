```python
import pandas as pd
import numpy as np
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#DataFrame with a mean of 5 and an SD of 1
d1 = pd.DataFrame({
    "size":np.random.normal(10,1,size=(100)),
    "type":"d1"})
d1.head()
```

```python
d1.describe()
```

```python
sns.distplot(d1['size'], rug=True).set(xlim=(0, 20))
```

```python
print(d1.mean(0))
print(d1.median(0))
```

```python
#DataFrame with a mean of 5 and an SD of 1
d2 = pd.DataFrame({
    "size":np.random.normal(10,3,size=(100)),
    "type":"d2"})
d2.describe()
```

```python
sns.distplot(d2["size"], rug=True).set(xlim=(0, 25))
```

```python
print(d2.mean(0))
print(d2.median(0))
d2.tail()
```

```python
extra = pd.DataFrame({"size":[14],"type":["extra"]})
df = pd.concat([d1, d2, extra])
sns.stripplot(x='size', data=df, y='type', jitter=True);

```
