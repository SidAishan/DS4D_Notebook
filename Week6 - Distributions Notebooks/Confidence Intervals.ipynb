```python
import pandas as pd
import numpy as np
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1a = pd.DataFrame({
    "value":np.random.normal(10,2,size=(50)),
    "condition":"a",
    "setup":"n=50,sd=2"})
```

```python
sns.barplot(x="setup",y="value",hue="condition",data=df1a)
```

```python
df1b = pd.DataFrame({
    "value":np.random.normal(15,2,size=(50)),
    "condition":"b",
    "setup":"n=50,sd=2"})
```

```python
sns.barplot(x="setup",y="value",hue="condition",data=pd.concat([df1a,df1b]))
```

```python
df2a = pd.DataFrame({"value":np.random.normal(10,10,size=(50)),"condition":"a","setup":"n=50,sd=10"})
df2b = pd.DataFrame({"value":np.random.normal(15,10,size=(50)),"condition":"b","setup":"n=50,sd=10"})
```

```python
sns.barplot(x="setup",y="value",hue="condition",data=pd.concat([df2a,df2b]))
```
